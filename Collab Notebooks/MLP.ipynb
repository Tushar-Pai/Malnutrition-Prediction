```python
import pandas as pd
df=pd.read_csv("newdataset.csv")
df
```

```python
df=df.drop(['Unnamed: 0'],axis=1)
```

```python
import numpy as np
```

```python
X= df[['BMDSTATS'	,'RIAGENDR',	'RIDAGEYR',	'BMXWT',	'BHXHT',	'BMXLEG',	'BOXARML',	'BMXARMC',	'BMXWAIST',	'BMXHIP']].values
X[0:10]
```

```python
y=df['OUTMAL'].values
y[0:10]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print('Train set :',X_train.shape,y_train.shape)
print('Test set :',X_test.shape,y_test.shape)
```

```python
from sklearn.neural_network import MLPClassifier
mlp=MLPClassifier(random_state=1, max_iter=500,activation='relu')
```

```python
mlp.fit(X_train,y_train)
```

```python
pred=mlp.predict(X_test)
```

```python
pred
```

```python
from sklearn import metrics

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay,classification_report
import matplotlib.pyplot as plt
print("MLP's accuracy is",metrics.accuracy_score(y_test,pred))
```

```python
cm=confusion_matrix(y_test,pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=[1,0])
disp.plot()
plt.show()
```

```python
print(classification_report(y_test,pred))
print("MLP'S accuracy",metrics.accuracy_score(y_test,pred))
```
